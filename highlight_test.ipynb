```python
import html
import random

import pandas as pd
from IPython.core.display import display, HTML
from pandas import DataFrame


# Prevent special characters like & and < to cause the browser to display something other than what you intended.
def html_escape(text):
    return html.escape(text)

text = "Ireland became a member of the European Union (EU) and joined " \
       "the single market on 1st January 1973. Before accession to the " \
       "bloc, Ireland had decades of an underachieving economy which was " \
       "heavily dependent on the UK."

# Remove duplicate words from text
seen = set()
result = []
for item in text.split():
    if item not in seen:
        seen.add(item)
        result.append(item)

#Create random sample weights for each unique word
weights = []
for i in range(len(result)):
    weights.append(random.random())

df_coeff = pd.DataFrame(
    {'word': result,
     'num_code': weights
    })

#Select the code value to generate different weights
word_to_coeff_mapping = {}
for row in df_coeff.iterrows():
    row = row[1]
    word_to_coeff_mapping[row[0]] = (row[1])

max_alpha = 0.8
highlighted_text = []
for word in text.split():
    weight = word_to_coeff_mapping.get(word)

    if weight is not None:
        highlighted_text.append('<span style="background-color:rgba(135,206,250,' + str(weight / max_alpha) + ');">' + html_escape(word) + '</span>')
    else:
        highlighted_text.append(word)

highlighted_text = ' '.join(highlighted_text)

display(HTML(highlighted_text))

df_data = ['<span style="background-color:rgba(135,206,250,0.9587091405951557);">Ireland</span> <span style="background-color:rgba(135,206,250,0.28972811548453253);">became</span> <span style="background-color:rgba(135,206,250,0.8377068384757411);">a</span> <span style="background-color:rgba(135,206,250,0.1615244105418165);">member</span> <span style="background-color:rgba(135,206,250,0.8879853712004395);">of</span> <span style="background-color:rgba(135,206,250,1.0704630544699745);">the</span> <span style="background-color:rgba(135,206,250,1.1512374637369527);">European</span> <span style="background-color:rgba(135,206,250,0.39925457702865275);">Union</span> <span style="background-color:rgba(135,206,250,1.2177224287753712);">(EU)</span> <span style="background-color:rgba(135,206,250,0.476563528266489);">and</span> <span style="background-color:rgba(135,206,250,0.26450713137144033);">joined</span> <span style="background-color:rgba(135,206,250,1.0704630544699745);">the</span> <span style="background-color:rgba(135,206,250,1.164527207351168);">single</span> <span style="background-color:rgba(135,206,250,0.7366579104812577);">market</span> <span style="background-color:rgba(135,206,250,0.4401804548956237);">on</span> <span style="background-color:rgba(135,206,250,0.908716557227561);">1st</span> <span style="background-color:rgba(135,206,250,0.9377022954488012);">January</span> <span style="background-color:rgba(135,206,250,1.0847752585924098);">1973.</span> <span style="background-color:rgba(135,206,250,0.6399194436991381);">Before</span> <span style="background-color:rgba(135,206,250,0.5250099577845724);">accession</span> <span style="background-color:rgba(135,206,250,1.1424686430053896);">to</span> <span style="background-color:rgba(135,206,250,1.0704630544699745);">the</span> <span style="background-color:rgba(135,206,250,0.9753674264717427);">bloc,</span> <span style="background-color:rgba(135,206,250,0.9587091405951557);">Ireland</span> <span style="background-color:rgba(135,206,250,0.17723162662331357);">had</span> <span style="background-color:rgba(135,206,250,0.3031627067607448);">decades</span> <span style="background-color:rgba(135,206,250,0.8879853712004395);">of</span> <span style="background-color:rgba(135,206,250,0.7174890165877466);">an</span> <span style="background-color:rgba(135,206,250,0.3556707816202598);">underachieving</span> <span style="background-color:rgba(135,206,250,0.500182110790163);">economy</span> <span style="background-color:rgba(135,206,250,0.7725935621525694);">which</span> <span style="background-color:rgba(135,206,250,1.0439461890065465);">was</span> <span style="background-color:rgba(135,206,250,0.724759082464077);">heavily</span> <span style="background-color:rgba(135,206,250,0.9263529215019369);">dependent</span> <span style="background-color:rgba(135,206,250,0.4401804548956237);">on</span> <span style="background-color:rgba(135,206,250,1.0704630544699745);">the</span> <span style="background-color:rgba(135,206,250,0.03499571384547373);">UK.</span>']
df = DataFrame(df_data, ['col'])
df

html = f'''
<table>
    <thead>
        <tr>
            <th>Entity</th>
            <th>Sentences</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td>James Bond</td>
            <td>
                <p>{highlighted_text}</p>
                <p>{highlighted_text}</p>
            </td>
        </tr>
    </tbody>
</table>
'''

display(HTML(html))
```
